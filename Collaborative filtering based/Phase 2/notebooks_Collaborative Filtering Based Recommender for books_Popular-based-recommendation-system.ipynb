## Collaborative based Recommendation System

```python
x = rating_with_name.groupby('User-ID').count()['Book-Rating'] > 200
user_vote_books = x[x].index
```

```python
user_vote_books
```

```python
filtered_rating = rating_with_name[
    rating_with_name['User-ID'].isin(user_vote_books)
]
```

```python
filtered_rating 
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
famous_books
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
pt = final_ratings.pivot_table(
    index='Book-Title',
    columns='User-ID',
    values='Book-Rating'
)
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # Find the index of the book
    index = np.where(pt.index == book_name)[0][0]

    # Get similarity scores for that book
    distances = similarity_scores[index]

    # Sort books by similarity (highest first)
    similar_items = sorted(
        list(enumerate(distances)),
        key=lambda x: x[1],
        reverse=True
    )[1:6]

    # Print recommended books
    for item in similar_items:
        print(pt.index[item[0]])
```

```python
recommend('1984')
```
